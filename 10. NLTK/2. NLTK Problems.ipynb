## Q1. From Gutenberg project, use 'austen-sense.txt' for the following task.

### i. Tokenize the data into words.

```python
#importing libraries
import nltk
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.corpus import gutenberg
import re
```

```python
#reading austen-sense text
text = gutenberg.sents(fileids='austen-Sense.txt')
arr=[]
#storing words into arr
for sentence in text:
    for word in sentence:
        arr.append(word)       
```

### ii. Remove all tokens which are stopwords or non alphanumeric characters.

```python
#stopwords in english
stop_words = set(stopwords.words('english'))
#removing stopwords
word_tokens = [x for x in arr if x.lower() not in stop_words]
```

```python
import string
#punctuations
punct = set(string.punctuation)
#removing punctuations from text
words = [x for x in word_tokens if x not in punct]
```

### iii. Convert all words to lowercase.

```python
#converting to lowercase
words_lowercase=[]
for word in words:
    words_lowercase.append(word.lower())
print(words_lowercase)
```

### iv. Apply and compare Porter Stemmer and Lancaster Stemmer on top 50 most frequent words. Choose the one that better suits your data and apply on the entire dataset.

```python
#count of each word
word_count={}
for word in words_lowercase:
    if word in word_count:
        word_count[word]+=1
    else:
        word_count[word]=1      
```

```python
#sorting the words in descending order based on count
sorted_wordcount=sorted(word_count.items(),key=lambda x:x[1],reverse=True)
top50 = sorted_wordcount[:50]
```

```python
#stemming words with porter stemmer for top50 most frequent words
from nltk.stem import PorterStemmer
ps = PorterStemmer()
ps_top50=[]
for w in top50:
    ps_top50.append(ps.stem(w[0])) 
print(ps_top50)    
```

```python
#stemming words with lancaster stemmer for top50 most frequent words
from nltk.stem import LancasterStemmer
ls=LancasterStemmer()
ls_top50=[]
for w in top50:
    ls_top50.append(ls.stem(w[0])) 
print(ls_top50)    
```

```python
#porter stemmer for all words
ps = PorterStemmer()
stemmed_words=[]
for w in words_lowercase:
    stemmed_words.append(ps.stem(w)) 
print(stemmed_words)    
```

for top50 most frequent words,stemmed words using porter stemmer are meaningful compared to lancaster stemmer.so,i used porter stemmer on the entire data

### Q2. From Gutenberg project, read the 'blake-poems.txt' as raw data. Perform following operations on the same:

### i. Tokenize the data into sentences.

```python
#reading raw data
data = gutenberg.raw('blake-poems.txt')
```

```python
#tokenize into sentences
sent_tokens = sent_tokenize(data)
```

### ii. For each tokenized sentence remove white-space characters.

```python
#removing whitespaces from each sentence
sent = [x.replace('\n',' ').replace('\t',' ').replace('\r',' ') for x in sent_tokens]
sent
```

### iii. For each tokenized sentence remove punctuation marks.

```python
#removing punctuations from each sentence
punc_removed = []
for sentence in sent
    new_sent = ""
    for char in sentence:
        if char not in punct:
            new_sent+=char
    punc_removed.append(new_sent)
```

### iv. Convert all words to lower-case. 

```python
#converting words to lowercase
lowercase_sent = [" ".join(word.lower().split()) for word in punc_removed]
```

### v. Remove 'stopwords' from the sentences. 

```python
#removing stopwords
stopwords_removed = []
for sentence in lowercase_sent:
    new_sent = []
    for word in sentence.split():
        if word not in stop_words:
            new_sent.append(word)
    stopwords_removed.append(" ".join(new_sent))
```

### vi. Remove all non-alpha numeric characters from the sentences.

```python
#removing nonalpha numeric characters from each sentence
non_alpha_num_removed = []
for sentence in stopwords_removed:
    new_sent = []
    for word in sentence.split():
        new_word =''
        for char in word:
            if char.isalnum():
                new_word+=char
        new_sent.append(new_word)
            
    non_alpha_num_removed.append(" ".join(new_sent))
non_alpha_num_removed
```
