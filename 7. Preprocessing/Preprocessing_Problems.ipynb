
	ii. Convert column three to numeric form by using any convenient mapping to integers.

ii. Eliminate the columns titled: 'date', all columns starting with 'sqft_*', 'statezip' and 'country'.

```python
# droping the given columns
df= df.drop(["date","statezip","country"],axis=1)
```

```python
# droping the columns starting with sqft_
df= df.drop(df.filter(regex='sqft_').columns,axis=1)
df
```

iii. Eliminate all columns if more than 50% of its values are set to 0. This is the cleaned dataframe. Display the shape of this cleaned dataframe.

```python
# finding total count
total_count = df["price"].count()
total_count
```

```python
#list to store the columns to remove
columns_to_remove = []
# finding  all columns with more than 50% of 0's
for col in df.columns:
    count = df[col].value_counts()
    if 0 in count:
        if count[0] > total_count/2:
            columns_to_remove.append(col)
columns_to_remove        
```

```python
# removing the columns
clean_df= df.drop(columns_to_remove,axis=1)
clean_df
```

```python
#showing the shape after the removal of cols
clean_df.shape
```

ix. Remove outliers based on the observations in (viii.).

```python
#Logic for removal of outliers
threshold = 0.077
for value in normalized_data['price']:
    #droping the rows containing outliers
    normalized_data = normalized_data.drop(normalized_data[normalized_data['price']>threshold].index)
    
df_out=pd.DataFrame(normalized_data['price'])
df_out.columns=['price']
```

```python
#boxplot after removing the outliers 
df_out.boxplot()
```

```python
#Logic for removal of outliers
thresholdLower = 0.2
thresholdUpper = 0.6
for value in normalized_data['bedrooms']:
    #droping the rows containing outliers
    normalized_data = normalized_data.drop(normalized_data[normalized_data['bedrooms'] > thresholdUpper].index)
    normalized_data = normalized_data.drop(normalized_data[normalized_data['bedrooms'] < thresholdLower].index)

#boxplot after removing the outliers 
df_out=pd.DataFrame(normalized_data['bedrooms'])
df_out.columns=['bedrooms']
df_out.boxplot()
```

```python
#Logic for removal of outliers
thresholdUpper = 0.6

for value in normalized_data['bathrooms']:
    #droping the rows containing outliers
    normalized_data = normalized_data.drop(normalized_data[normalized_data['bathrooms'] > thresholdUpper].index)
#boxplot after removing the outliers       
df_out=pd.DataFrame(normalized_data['bathrooms'])
df_out.columns=['bathrooms']
df_out.boxplot()
```

```python
#Logic for removal of outliers

thresholdLower = 0.05
for value in normalized_data['yr_built']:
    #droping the rows containing outliers
    normalized_data = normalized_data.drop(normalized_data[normalized_data['yr_built'] < thresholdLower].index)
#boxplot after removing the outliers 
df_out=pd.DataFrame(normalized_data['yr_built'])
df_out.columns=['yr_built']
df_out.boxplot()
```

```python
#Logic for removal of outliers
thresholdupper = 0.4
for value in normalized_data['condition']:
    #droping the rows containing outliers
    normalized_data = normalized_data.drop(normalized_data[normalized_data['condition'] > thresholdupper].index)
#boxplot after removing the outliers        
df_out=pd.DataFrame(normalized_data['condition'])
df_out.columns=['condition']
df_out.boxplot()
```
