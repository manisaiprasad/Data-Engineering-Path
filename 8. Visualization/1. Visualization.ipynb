# Matplotlib.pyplot

*   Object name is generally defined as **plt**.
*   Convenient interface to the matplotlib object-oriented plotting library.
*   Modeled closely after Matlab(TM).



*   Matplotlib plots the datapoints on a **figure**.
*   A **figure** has one or more **axes**.
*   **Axes**:
    *   x-y coordinates for 2D plots
    *   x-y-z coordinates for 3D plots
    *   theta-r coordinates for polar plots
*   We can also draw the datapoints using **plt.plot()** function and use **plt.show()** to display the plot.


## Table

```python
import matplotlib.pyplot as plt 

plt.style.use('ggplot')
fig = plt.figure(dpi=120)
ax = fig.add_subplot(1,1,1)

table_data=[['E001', 'M', 34, 123, 'Normal', 350], 
        ['E002', 'F', 40, 114, 'Overweight', 450], 
        ['E003', 'F', 37, 135, 'Obesity', 169], 
        ['E004', 'M', 30, 139, 'Underweight', 189], 
        ['E005', 'F', 44, 117, 'Underweight', 183], 
        ['E006', 'M', 36, 121, 'Normal', 80], 
        ['E007', 'M', 32, 133, 'Obesity', 166], 
        ['E008', 'F', 26, 140, 'Normal', 120], 
        ['E009', 'M', 32, 133, 'Normal', 75], 
        ['E010', 'M', 36, 133, 'Underweight', 40]]

table = ax.table(cellText = table_data, loc='center')
table.set_fontsize(14)
ax.axis('off')

plt.show()
plt.close()
```

## Histogram

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

# Splitting the features and the labels
iris_features = iris.data
iris_labels = iris.target
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width']
iris_species = ["Setosa", "Versicolour", "Virginica"]

dataset = pd.DataFrame(iris_features, columns=names)
dataset.head()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.gca()  # Alternative to plt.subplots for simpler plots

# Plot histograms
dataset.hist(ax=ax, color='b',alpha=0.7, rwidth=0.85)
plt.tight_layout()
plt.show()
plt.close()
```

## Pie-Chart

```python
import numpy as np
fig, ax = plt.subplots(figsize=[7,7])        # Create a blank canvas, 7inches wide by 7inches tall.
                                             # Keeping it square looks best for a pie chart
ax.set_title("Proportion of each species")   # Add title to the top of the plot
pie = ax.pie(np.bincount(iris_labels), labels=iris_species, autopct='%1.1f%%') #'bincount' counts how many times each integer shows up. Y has values from [0,1,2] 
fig.savefig("clean_pie.png")                 # Save the figure
plt.show()
plt.close(fig)                               # Close the canvas
```

```python
explode = (0.1, 0, 0)  # explode 1st slice

# Define custom colors
colors = ['cyan', 'lightblue', 'lightgreen']

# Plot
fig, ax = plt.subplots(figsize=[7,7])
ax.set_title("Proportion of each species")
pie = ax.pie(np.bincount(iris_labels), explode=explode, labels=iris_species, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
plt.close(fig)
```

```python
# Add legend to Pie-chart
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=[7,7])
ax.set_title("Proportion of each species")
patches, texts = plt.pie(np.bincount(iris_labels), explode=explode, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels=iris_species, loc="best")
# plt.axis('equal')
plt.tight_layout()
plt.show()
plt.close(fig)
```

## Scatter-plot

```python
colors = ['blue', 'red', 'green']

fig = plt.figure(figsize = (8,6))
for i in range(len(colors)):
    px = iris_features[:,0][iris_labels==i]
    py = iris_features[:,1][iris_labels==i]
    plt.scatter(px, py, c=colors[i])
plt.legend(iris_species)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
plt.close(fig)
```

## Bubble plot

```python

np.random.seed(1234)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (40 * np.random.rand(N))**2  # area for different points.    

fig = plt.figure(figsize = (8,6))
plt.style.use('dark_background')
plt.scatter(x, y, s=area, c=colors, alpha=0.6)
plt.show()
plt.close()
```

## Waffle chart

```python
# python program to generate Waffle Chart 
  
# importing all necessary requirements
from pywaffle import Waffle 
  
# creation of a dataframe 
data ={'laptop': ['Dell', 'HP', 
                 'Mac', 'Lenovo', 'Acer',
                  'Asus', 'Samsung', 'Sony'], 
       'units_sold': [25, 45, 5, 12, 27, 10, 3, 1]  # Units sold per year
     } 
  
df = pd.DataFrame(data) 
  
# To plot the waffle Chart 
plt.style.use('dark_background')
fig = plt.figure(
    figsize=(8,6),
    FigureClass = Waffle, 
    rows = 8, 
    values = df.units_sold, 
    labels = list(df.laptop) 
)
plt.show()
plt.close()
```

## WordCloud

```python
# Python program to generate WordCloud 
  
# importing all necessary modules 
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
  
# Reads 'Youtube05-Shakira.csv' file  
df = pd.read_csv(r"Youtube05-Shakira.csv", encoding ="latin-1") 
  
comment_words = ''
stopwords = set(STOPWORDS)
  
# iterate through the csv file 
for val in df.CONTENT: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8)) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
plt.close()
```
